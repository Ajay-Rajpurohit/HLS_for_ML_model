### Issues Solved during Installation of Package Dependencies:

#### Problem:

Some package dependencies required specific versions which differed from the versions previously installed on our system.

#### Example:

- Package: TensorFlow
- Earlier Version: 2.16.0
- Requirement: <=2.14.0

#### Solution:

Noticed that the installed version of TensorFlow (2.16.0) did not meet the requirement (<=2.14.0) specified by the package dependencies.

#### Resolve the Issue:

- Uninstall the existing version of TensorFlow:
pip uninstall tensorflow

- Install the specific version required (2.14.0):
pip install tensorflow==2.14.0

#### Outcome:

Successfully resolved the version compatibility issue for TensorFlow and ensured that the package dependencies were met according to the specified requirements.

### Issue: C Synthesis Report Not Found

#### Problem:

When attempting to run the model on hls4ml in a Windows system, encountered an error stating that the C synthesis report was not found.

#### Solution:

1. <b>Troubleshoot in Windows:</b>
Initially attempted to resolve the issue within the Windows environment, but encountered difficulties locating the C synthesis report.
2. <b>Switch to Linux Based System (Ubuntu):</b>
Opted to switch the operating system to Linux-based Ubuntu.
Reran the model on hls4ml within the Ubuntu environment.
3. <b>Resolution:</b>
Successfully ran the model on hls4ml in Ubuntu without encountering the "C synthesis report not found" error.
The issue was resolved by transitioning to a Linux-based system, where the required C synthesis report was generated and accessible.

Default clock set to 5 nanoseconds, conflicting with Vivado HLS generated code which uses 10 nanoseconds.
- <b>Solution:

Adjusted clock period variable in build_prj.tcl to 10 nanoseconds to match Vivado HLS.
- <b>Outcome:

Clock period aligned with Vivado HLS, ensuring consistency in design.

- This setting, when enabled (true), likely configures the tool to fully register DSP blocks. DSP blocks (Digital Signal Processing) are specialized hardware units on FPGAs used for high-speed arithmetic operations, such as those found in signal processing and deep learning computations.
